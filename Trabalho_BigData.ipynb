```python
# Vamos limpar esses campos com os valores NULL usando a função "dropna()"
netflix_no_null = netflix.dropna().copy()
netflix_no_null
```

```python
# Agora vamos fazer nossas primeiras analises do arquivo
# Agora, identificaremos a diferença em quantidade de filmes e shows de TV, mostraremos em um gráfico para melhor visualização
# Agrupa todos os items de uma coluna e mostra a quantidade de cada.
netflix.groupby('Category').Category.count()
```

```python
# O Gráfico abaixo mostra a diferênça de quantidade de filmes e show de TV.
sns.countplot(x=netflix['Category'])
```

```python
#Faremos um filtro de todos os filmes estreados no ano de 2021.
#Faremos uma nova coluna com a data formatada (datetime format) e em seguida criaremos uma coluna apenas com o ano (atributo year).
netflix['Date_Formated'] = pd.to_datetime(netflix['Release_Date'])
```

```python
# Criando uma nova coluna apenas com o ano de realização do filme.
netflix['Year'] = netflix['Date_Formated'].dt.year
```

```python
# Filtrando as duas condições iniciais: Ser filme de 2021.
netflix [(netflix['Category'] == 'Movie') & (netflix['Year'] == 2021)]
```

```python
# Agora vamos analisar os títulos de todos os filmes realizados no Brasil
# Caso queira analisar títulos de filmes realizados em outros países, voçê apenas modifica o país
# Na posição 'Title', pois queremos apenas os títulos
netflix[(netflix['Category'] == 'Movie') & (netflix['Country'] == 'Brazil')]['Title']
```

```python
# Aqui listaremos um comando para verificar a produção de filmes brasileiros
# Quantidade de produções brasileiras
netflix[(netflix['Category'] == 'Movie') & (netflix['Country'] == 'Brazil')]['Country'].value_counts()
```

```python
# Top 10 filmes mais assistidos
netflix = netflix.sort_values(by='Rating', ascending=False)
top_10_filmes = netflix.head(10)
display(top_10_filmes[['Title']])
```

```python
#Agora analisaremos o contexto de produções referente a todos os países
#Plotagem do gráfico para melhor visualização.
netflix['Country'].value_counts().head(15).plot.barh(title='País x Qtd. Produções')
```

```python
#Agora vamos verificar os diretores.
#Anteriormente, a coluna diretores apresentou grande quantidade de valores nulos.
#Vamos puxa a lista.

#Função de agregação (count) agrupado pelos diretores.

count_directors = netflix_no_null['Director'].value_counts()
count_directors
```

```python
# Listamos a quantidade de diretores sem tratamento
len(pd.unique(netflix['Director']))
```

```python
# Lista de diretores excluindo os valores Nulos
sum_directors = len(pd.unique(netflix_no_null['Director']))
sum_directors
```

```python
from pandas.core.arraylike import maybe_dispatch_ufunc_to_dunder_op
# Lista de quantidade de produção
sum_productions = len(pd.unique(netflix_no_null['Show_Id']))
# Aqui está a Média de produção por diretor
mean_director = sum_productions / sum_directors
mean_director
```

```python
# Usando a biblioteca numpy
# Média de produção por diretor usando a bibliteca Numpy
np_mean = np.average(netflix_no_null['Director'].value_counts())
np_mean
```

```python
n1 = np.std(netflix_no_null['Director'].value_counts(), ddof=1)
n1 # Valor do desvio padrão com grau de liberdade igual a 1 (ddof = 1)
```

```python
cv = (n1 / np_mean) * 100 # Coeficiente de variação.
cv
```

```python
# Agora vamos avaliar a discrepância de produção entre os diretores, usando uma nova biblioteca "MATPLOTIB"
# O comando plt.tilte gera um grafico e .title colocamos o título nele
plt.title('Quantidade de produções por cada diretor')
# O comando "SCATTER" vai plotar pontos azuis no gráfico e sua quantidade
plt.scatter(x=range(sum_directors), y = count_directors, s = 5)
# O comando "HLINES" plota uma linha vermelha no gráfico para representar a média
plt.hlines(y=np_mean, xmin=0, xmax=sum_directors, colors='red')
# Nomeando ìndice X do gráfico
plt.xlabel("Índice do diretor")
# Nomeando Índice Y do gráfico
plt.ylabel('Quantidade de produções')
# Aqui definimos a aultura das linhas
plt.grid(linewidth = 0.6)
# Agora solicitamos mostrar o gráfico
plt.show()
```

```python
# Agora vamos ver os filmes com maior duração
netflix.Duration.unique()
```

```python
# Vamos precisar separar o números das unidades
# E tbm não será necessário separar os filmes dos TV Show's, pois não tem nenhum com mais temporadas que a quantodade de minutos
# Abaixo o comando para separar as strind em duas (Tempo e Unidade)
netflix[['Minutes','Unit']] = netflix.Duration.str.split(' ', expand = True)
# Agora vamos converte um pandas object para uma string
netflix['Minutes'] = netflix['Minutes'].astype(str).astype(int)
# Vamos buscar o valor máximo de duração dos filmes
max_duration = netflix["Minutes"].astype(str).astype(int)
# Agora vamos mostrar o filme com maior duração
netflix[netflix["Minutes"] == max_duration]
```

```python
# Aqui vamos fazer o oposto, buscaremos filmes com menor duração
# Abaixo vamos filtrar apenas filmes
netflix_movies = netflix[netflix["Category"] == "Movie"]
# Solicitamos a pesquisa, apenas com 3 cabeçalho por filme
netflix_movies.head(3)
```

```python
# Agora vamos mostrar o com o menor tempo de duração
netflix_movies[netflix_movies["Minutes"] == netflix_movies["Minutes"].min()]
```

```python
# Agora vamos mostrar o com o maior tempo de duração
netflix_movies[netflix_movies["Minutes"] == netflix_movies["Minutes"].max()]
```

```python
# Agora vamos Ordenar os filmes e TV show pelo ano Publicado
#False é decrescente e True é crescente
netflix.sort_values(by='Release_Date', ascending= False)
```

```python
# Agora vamos selecionar a categoria de Filmes e tipo em geral
# Vamos filtrar apenas filmes de comédia
netflix[(netflix["Category"] == "Movie") & (netflix["Type"] == "Comedies")]
```

```python
# Filtrar filmes que contém drama (possui outros gêneros mas o drama está incluso)
netflix[(netflix['Category'] == 'Movie') & (netflix['Type'].str.contains('Drama'))]
```
